```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

file_path = 'C:\\Users\\stann\\OneDrive\\Documents\\datasets\\movie.csv'
movie_data = pd.read_csv(file_path)

print("First few rows of the dataset:")
print(movie_data.head())
```

```python
print("\nGeneral Statistics:")
print(movie_data.describe())
```

```python
movie_data.fillna(0, inplace=True)
```

```python
plt.figure(figsize=(12, 6))
sns.countplot(y='genre', data=movie_data, order=movie_data['genre'].value_counts().index)
plt.title('Distribution of Movies by Genre')
plt.xlabel('Number of Movies')
plt.ylabel('Genre')
plt.show()
```

```python
plt.figure(figsize=(12, 6))
sns.boxplot(x='rating', y='score', data=movie_data)
plt.title('IMDb Scores by Movie Rating')
plt.xlabel('Movie Rating')
plt.ylabel('IMDb Score')
plt.show()
```

```python
plt.figure(figsize=(12, 6))
sns.scatterplot(x='budget', y='gross', data=movie_data)
plt.title('Budget vs. Gross Earnings')
plt.xlabel('Budget (in dollars)')
plt.ylabel('Gross Earnings (in dollars)')
plt.show()
```

```python
plt.figure(figsize=(12, 6))
sns.countplot(x='year', data=movie_data, palette='viridis')
plt.title('Number of Movies Released Each Year')
plt.xlabel('Year')
plt.ylabel('Number of Movies')
plt.xticks(rotation=90)
plt.show()
```

```python
top_directors = movie_data['director'].value_counts().head(10)
plt.figure(figsize=(12, 6))
sns.barplot(x=top_directors.values, y=top_directors.index, palette='magma')
plt.title('Top 10 Directors by Number of Movies')
plt.xlabel('Number of Movies')
plt.ylabel('Director')
plt.show()
```

```python
top_stars = movie_data['star'].value_counts().head(10)
plt.figure(figsize=(12, 6))
sns.barplot(x=top_stars.values, y=top_stars.index, palette='plasma')
plt.title('Top 10 Stars by Number of Movies')
plt.xlabel('Number of Movies')
plt.ylabel('Star')
plt.show()
```
